##### Visualize the change in SDG index score for Afghanistan over time by a histogram

```python
# drop useless columns
df.drop(df.index[4:21], axis = 0, inplace = True)
```

```python
# drop useless rows
df.drop(df.iloc[22:4140].index, inplace = True)
```

```python
# visualize the change in SDG index score for Afghanistan over time by a histogram
import matplotlib.pyplot as plt
column_name = 'sdg_index_score'
df[column_name].hist(bins=10, grid=False, xlabelsize=12, ylabelsize=12)
plt.xlabel(column_name, fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Histogram of' + '' + column_name, fontsize=15)
plt.show()
```

### Conclusions
---
1. The mean of the SDG index score for Afghanistan is 48.6 between 2000 and 2022. This seems to be a low score but this could have been due to many outliers.
2. The variance of the SDG index score for Afghanistan is 39.9. This indicates that there are many outliers that have caused the mean to be lower than 50.
3. The histogram shows that the SDG index score for Afghanistan is skewed to the left. This further validates that there are many outliers.
